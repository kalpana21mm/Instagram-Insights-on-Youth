*Bubble chart: Desired changes*

*Purpose:  To visually represent the relationship between multiple variables (such as frequency and impact of desired changes), with the bubble size indicating the importance or magnitude of each change, helping to identify key areas users want to see altered.*

```python
import matplotlib.pyplot as plt
from collections import Counter
desired_changes_text = ' '.join(df['Desired changes'].dropna()) #Count the frequency of words/phrases in the 'Desired changes' column
word_counts = Counter(desired_changes_text.split())
words, counts = zip(*word_counts.items()) # Prepare data
sizes = [count * 100 for count in counts]  # Bubble size proportional to frequency
plt.figure(figsize=(10, 5))
plt.scatter(words, [1] * len(words), s=sizes, alpha=0.6, color='purple', edgecolors="w", linewidth=0.5)
plt.title('Bubble Chart of Desired Changes', fontsize=16)
plt.xticks(rotation=90, ha='right', fontsize=10)  # Rotate x-axis labels for readability
plt.yticks([])  # Hide y-axis
plt.tight_layout()
plt.show()
```

*Interpretation- Fewer respondents than expected prioritize privacy controls, content moderation, reducing ads, and focusing on user well-being, suggesting these concerns may not be as significant as anticipated.*

*Guage: Experience rating*

*Purpose: To provide a clear, intuitive visual representation of user satisfaction, showing the overall rating on a scale and highlighting the level of experience, whether it's positive, neutral, or negative.*

```python
import matplotlib.pyplot as plt
import numpy as np
mean_experience = df['Experience rating'].mean() # Calculate the mean experience rating (scale of 1 to 5)
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'projection': 'polar'}) # Create a polar plot for the gauge
ax.set_theta_offset(np.pi / 2)  # Start the gauge at the top
ax.set_theta_direction(-1)      # Gauge direction (clockwise)
sections = [1, 2, 3, 4, 5] # Define the sections and colors for a 1 to 5 rating scale
colors = ['red', 'orange', 'yellow', 'lightgreen', 'green']
for i in range(len(sections) - 1): # Plot the sections of the gauge
    ax.barh(0, width=np.pi / 4, height=1, left=(i * np.pi / 4), color=colors[i], edgecolor='black')
pointer_position = ((mean_experience - 1) / 4) * 2 * np.pi # Plot the mean experience rating as a pointer
ax.barh(0, width=0.05, height=1.5, left=pointer_position, color='black')
ax.set_yticks([]) # Set the tick labels with slight offset for visibility
tick_positions = np.linspace(0, 2 * np.pi, len(sections) + 1)[:-1]  # Exclude the last position to avoid overlap
ax.set_xticks(tick_positions)
ax.set_xticklabels(['1', '2', '3', '4', '5'], fontsize=12)
ax.set_title('Experience Rating Gauge (1 to 5 Scale)', fontsize=14) # Add title and formatting
plt.tight_layout()
plt.show()
```

*Interpretation- All age groups rate Instagram similarly, suggesting a consistent user experience across different demographics.*

*Area plot: Need for more protection*

*Purpose: To show cumulative trends, helping to understand whether the demand for more protection is increasing or decreasing over time, allowing for better response strategies.*

```python
import pandas as pd
import matplotlib.pyplot as plt
protection_counts = df['Need for more protection'].value_counts(normalize=True) # Calculate the proportions for "Yes" and "No" responses
yes_no = ['Yes', 'No']
values = [protection_counts.get('Yes', 0), protection_counts.get('No', 0)]
protection_df = pd.DataFrame({'Response': yes_no, 'Proportion': values})
plt.figure(figsize=(8, 5))
plt.fill_between(protection_df['Response'], protection_df['Proportion'], color='peachpuff', alpha=0.6)
plt.plot(protection_df['Response'], protection_df['Proportion'], color='olive', marker='o')
plt.title('Need for More Protection on Social Media', fontsize=14) # Formatting
plt.xlabel('Response', fontsize=12)
plt.ylabel('Proportion', fontsize=12)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

*Interpretation- All age groups support more protective measures for youth on social media, with 18-22 showing the most support and 28-32 the least, but differences are not significant.*
